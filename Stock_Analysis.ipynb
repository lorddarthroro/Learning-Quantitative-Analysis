```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

data = yf.download(["GOOGL"], start="2000-01-01", end=end_date)

df = data['Adj Close'].reset_index()


# df['Date'] = pd.to_datetime(df['Date'])

# Set 'date' as the index

df.set_index('Date', inplace=True)

# Train-test split
train = df.iloc[:-3]  
test = df.iloc[-3:]   
# 0.1325 and -0.0451, respectively.
# Fit ARIMA model
model = ARIMA(train, order=(0, 1, 0))  # Example order, adjust as needed
model_fit = model.fit()

# Forecast
forecast = model_fit.forecast(steps=len(test))

# Evaluate forecast
mse = mean_squared_error(test, forecast)
rmse = mse ** 0.5
print("RMSE:", rmse)
```

```python
# Forecast for the next 365 days
forecast = model_fit.forecast(steps=365)

# Create a DataFrame for the forecasted values
forecast_index = pd.date_range(start=df.index[-1], periods=365)  # Adjust periods to include the next 365 days
forecast_df = pd.DataFrame(forecast)
forecast_df = forecast_df.set_index(forecast_index)
# Add forecasted values to the original DataFrame
combined_df = pd.concat([df, forecast_df])

# Plot original and forecasted values
plt.figure(figsize=(10, 6))
plt.plot(combined_df.index, combined_df['Adj Close'], label='Actual')
plt.plot(combined_df.index, combined_df['predicted_mean'], label='Forecast', linestyle='--')
plt.title('DKNG Stock Price Forecast')
plt.xlabel('Date')
plt.ylabel('Adj Close Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import yfinance as yf
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt

# Download data
end_date = "2024-03-01"  # Assuming current date for the end date
data = yf.download(["GOOGL"], start="2020-01-01", end=end_date)

# Extract 'Adj Close' column and reset index
df = data['Adj Close'].reset_index()

# Set 'Date' as the index
df.set_index('Date', inplace=True)

# Fit SARIMAX model
model = ARIMA(df, order=(1, 1, 1))  # Example order, adjust as needed
model_fit = model.fit()

# Check model summary
print(model_fit.summary())

# Forecast
forecast = model_fit.forecast(steps=365)

# Create a DataFrame for the forecasted values
forecast_index = pd.date_range(start=df.index[-1], periods=365)  # Adjust periods to include the next 365 days
forecast_df = pd.DataFrame(forecast)
forecast_df = forecast_df.set_index(forecast_index)

combined_df = pd.concat([df, forecast_df])

# Plot original and forecasted values
plt.figure(figsize=(10, 6))
plt.plot(combined_df.index, combined_df['Adj Close'], label='Actual')
plt.plot(combined_df.index, combined_df['predicted_mean'], label='Forecast', linestyle='--')
plt.title('GOOGL Stock Price Forecast')
plt.xlabel('Date')
plt.ylabel('Adj Close Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import yfinance as yf
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt

# Download data
end_date = "2024-03-01"  # Assuming current date for the end date
data = yf.download(["GOOGL"], start="2020-01-01", end=end_date)

# Extract 'Adj Close' column and reset index
df = data['Adj Close'].reset_index()

# Set 'Date' as the index
df.set_index('Date', inplace=True)

# Fit SARIMA model
model = SARIMAX(df, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))  # Example orders, adjust as needed
model_fit = model.fit()

# Check model summary
print(model_fit.summary())

# Forecast
forecast = model_fit.forecast(steps=365)

forecast_index = pd.date_range(start=df.index[-1], periods=365)  # Adjust periods to include the next 365 days
forecast_df = pd.DataFrame(forecast)
forecast_df = forecast_df.set_index(forecast_index)

# Plot original and forecasted values
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Adj Close'], label='Actual')
plt.plot(forecast_df.index, forecast_df['predicted_mean'], label='Forecast', linestyle='--')
plt.title('GOOGL Stock Price Forecast with SARIMA')
plt.xlabel('Date')
plt.ylabel('Adj Close Price')
plt.legend()
plt.grid(True)
plt.show()
```
